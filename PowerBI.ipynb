**Interview Queastion**

What is the Difference between **New Measure** and creating a **New Column**?

Ans:

**New Measure** This is done for the Analysis of the Data.. There is no change done to the Original Table. It gives the aggregation Value. It always gives a Single value.
Example : SUM(),Average()

**New Column** Creating New Column is doing changes to the  Original Table ..it gives multiple records.If a Table contains 1000records, the **New Column** contains 1000 records.

Example:

Product Name   Product Quantity   Price per Unit   Sales_value

Ice cream        5                   rs 80           80*5 =400

Chocolate        8                   rs 60           60*8 =480




Here **Sales_Value**  is the New column that is created and not **New Measure** because it contains multiple values.
